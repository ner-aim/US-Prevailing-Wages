```python
# Filling the new column income_group

for i in range(len(result['PWD_WAGE_RATE'])):
    if result['PWD_WAGE_RATE'].iloc[i] >=0 and result['PWD_WAGE_RATE'].iloc[i]<=50000:
        result['income_group'].iloc[i] = '0-50k'
    elif result['PWD_WAGE_RATE'].iloc[i] >50000 and result['PWD_WAGE_RATE'].iloc[i]<=100000:
        result['income_group'].iloc[i] = '50k-100k'
    elif result['PWD_WAGE_RATE'].iloc[i] >100000 and result['PWD_WAGE_RATE'].iloc[i] <=200000:
        result['income_group'].iloc[i] = '100k-200k'
    elif result['PWD_WAGE_RATE'].iloc[i] >200000 and result['PWD_WAGE_RATE'].iloc[i] <=300000:
        result['income_group'].iloc[i] = '200k-300k'
    elif result['PWD_WAGE_RATE'].iloc[i] >300000 and result['PWD_WAGE_RATE'].iloc[i] <=400000:
        result['income_group'].iloc[i] = '300k-400k'
    elif result['PWD_WAGE_RATE'].iloc[i] >400000 and result['PWD_WAGE_RATE'].iloc[i] <=500000:
        result['income_group'].iloc[i] = '400k-500k'
    elif result['PWD_WAGE_RATE'].iloc[i] >500000:
        result['income_group'].iloc[i] = 'above 500k'
```
